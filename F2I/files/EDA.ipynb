```python
%pip install pandas
%pip install matplotlib
%pip install seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/Users/sasanksasi/Downloads/project/VertexAi/200_companies_sample.csv'
df = pd.read_csv(file_path)
```

```python
print(df.info())
print(df.describe())
print(df.head())
```

```python
df.dropna(axis=1, how='all', inplace=True)
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['number_of_investments'].dropna(), bins=30, kde=True)
plt.title('Distribution of Number of Investments')
plt.xlabel('Number of Investments')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of the number of exits
plt.figure(figsize=(10, 6))
sns.histplot(df['number_of_exits'].dropna(), bins=30, kde=True)
plt.title('Distribution of Number of Exits')
plt.xlabel('Number of Exits')
plt.ylabel('Frequency')
plt.show()

# Plot the number of companies by investor type
plt.figure(figsize=(14, 8))
sns.countplot(y='investor_type', data=df, order=df['investor_type'].value_counts().index)
plt.title('Number of Companies by Investor Type')
plt.xlabel('Count')
plt.ylabel('Investor Type')
plt.show()

# Plot the number of companies by country
plt.figure(figsize=(14, 8))
sns.countplot(y='country', data=df, order=df['country'].value_counts().index)
plt.title('Number of Companies by Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()

# Plot the number of companies by industry
plt.figure(figsize=(14, 8))
industries = df['industries'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y=industries, order=industries.value_counts().index)
plt.title('Number of Companies by Industry')
plt.xlabel('Count')
plt.ylabel('Industry')
plt.show()
```
